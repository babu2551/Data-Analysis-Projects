```python
df = df.drop(columns=['datetime'])
display(df.head())
```

```python
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
```

```python
df.fillna(0, inplace=True)
```

## Data Visualization

```python
sales_by_coffee = df.groupby('coffee_name')['money'].sum().sort_values(ascending=False)

plt.figure(figsize=(5, 4))
sns.barplot(x=sales_by_coffee.index, y=sales_by_coffee.values, palette="inferno" )
plt.title('Total Money Spent per Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Total Money Spent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 4))
plt.pie(sales_by_coffee.values, labels=sales_by_coffee.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("inferno" , len(sales_by_coffee)))
plt.title('Percentage of Total Money Spent per Coffee Type')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
cash_type_counts = df['cash_type'].value_counts()

plt.figure(figsize=(5, 3))
sns.barplot(x=cash_type_counts.index, y=cash_type_counts.values, palette='PuRd')
plt.title('Count of Transactions by Cash Type')
plt.xlabel('Cash Type')
plt.ylabel('Number of Transactions')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(5, 4))
sns.lineplot(x=sales_by_coffee.index, y=sales_by_coffee.values, palette='viridis', markers='o')
plt.title('Total Money Spent per Coffee Type')
plt.xlabel('Coffee Type')
plt.ylabel('Total Money Spent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_by_year = df.groupby('year')['money'].sum().sort_values(ascending=False)

plt.figure(figsize=(5, 4))
plt.bar(sales_by_year.index, sales_by_year.values, color='skyblue')
plt.title('Total Money Spent per Year')
plt.xlabel('Year')
plt.ylabel('Total Money Spent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
sales_by_month = df.groupby('month')['money'].sum().sort_values(ascending=False)

plt.figure(figsize=(5, 4))
plt.bar(sales_by_month.index, sales_by_month.values, color='blue')
plt.title('Total Money Spent per Year')
plt.xlabel('Month')
plt.ylabel('Total Money Spent')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
