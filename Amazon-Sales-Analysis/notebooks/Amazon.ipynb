```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('sample_data/amazon.csv')
```

```python
df.head(2)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df.isna()
```

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)

df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
```

```python
df.head(2)
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '', regex=False)
df['discounted_price'] = df['discounted_price'].str.replace(',', '', regex=False)
df['discounted_price'] = df['discounted_price'].astype(float)
```

```python
df['actual_price'] = df['actual_price'].str.replace('₹', '', regex=False)
df['actual_price'] = df['actual_price'].str.replace(',', '', regex=False)
df['actual_price'] = df['actual_price'].astype(float)
```

```python
df.head(2)
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)
df['rating_count'] = df['rating_count'].astype(float)
```

```python
df.head(2)
```

```python
df.duplicated().sum()
```

```python
highest_rated = df.groupby('category')['rating'].mean().sort_values(ascending=False)
highest_rated.head(10)
```

```python
most_review_product = df.groupby('product_name')['rating_count'].sum().sort_values(ascending=False)
most_review_product.head(10)
```

```python
df[['discount_percentage','rating']].corr()
```

```python
plt.scatter(df['discount_percentage'], df['rating'])
plt.xlabel('Discount Percentage')
plt.ylabel('Rating')
plt.title('Discount vs Rating')
plt.grid(True)
plt.show()
```

```python
df['saving'] = df['actual_price'] - df['discounted_price']
df.head(10).sort_values(by='saving', ascending=False)
```

```python
av_price = df['actual_price'].mean()
av_discounted_price = df['discounted_price'].mean()
plt.bar(['Actual Price', 'Discounted Price'], [av_price, av_discounted_price], color=['blue', 'green'])
plt.xlabel('Price Type')
plt.ylabel('Average Price')
plt.title('Average Actual Price vs Average Discounted Price')
plt.show()
```

```python
expensive_products = df.nlargest(10, 'discounted_price')
print(expensive_products[['product_name','actual_price']])
```

```python
average_rating = df.groupby('category')['rating'].mean()
average_rating.head(10)
```

```python
average_discount = df.groupby('category')['discounted_price'].mean()
average_discount.head(10)
```

```python
df['rating'].describe()
```

```python
plt.hist(df['rating'], bins=10)
plt.xlabel('Rating')
plt.ylabel('Number of Products')
plt.title('Rating Distribution')
plt.show()
```

```python
df['rating'].value_counts().sort_index()
```

```python
df['value_score'] = (
    df['rating'] *
    df['discount_percentage']
) / df['discounted_price']

df.head(5)
```

```python
top_value = df[
    ['product_name',
     'rating',
     'discount_percentage',
     'discounted_price',
     'value_score']
].sort_values(
    'value_score',
    ascending=False
)

top_value.head(10)
```
