```python
release_count = data['release_year'].value_counts().sort_index()
plt.figure(figsize=(12,5))

plt.bar(
    release_count.index,
    release_count.values,
    color='red'
    )

plt.title('Netflix Content Release Trend')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend()

plt.show()
```

```python
country_count = data['country'].value_counts().head(10)

plt.figure(figsize=(12,5))
plt.bar(
    country_count.index,
    country_count.values
    )

plt.title('Top 10 Countries on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.legend()

plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.pie(
    country_count.values,
    labels=country_count.index,
    autopct='%1.1f%%'
    )
# plt.legend()
plt.title('Top 10 Countries on Netflix')
plt.show()
```

```python
type_count = data['type'].value_counts()
plt.figure(figsize=(12,5))
plt.pie(
    type_count.values,
    labels=type_count.index,
    autopct='%1.1f%%'
    )
plt.title('Distribution of Content Types on Netflix')
plt.show()
```

```python
rating_count = data['rating'].value_counts()
plt.figure(figsize=(12,5))
plt.plot(
    rating_count.index,
    rating_count.values,
    marker='o',
    linestyle='-',
    color='b'
    )
plt.xlabel('Rating')
plt.ylabel('Count')
plt.title('Distribution of Ratings on Netflix')
```

```python
movie_data = data[data['type'] == 'Movie']

duration = movie_data['duration'].str.replace(' min','')
duration = duration.astype(int)

plt.figure(figsize=(10,5))

plt.hist(duration,
         bins=20)

plt.title('Movie Duration Distribution')
plt.xlabel('Minutes')
plt.ylabel('Frequency')

plt.show()
```

```python
india = data[data['country'] == 'India']
usa = data[data['country'] == 'United States']

india_count = india['release_year'].value_counts().sort_index()
usa_count = usa['release_year'].value_counts().sort_index()

plt.figure(figsize=(12,6))

plt.plot(india_count.index,
         india_count.values,
         label='India')

plt.plot(usa_count.index,
         usa_count.values,
         label='USA')

plt.title('India vs USA Netflix Releases')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')

plt.legend()
plt.grid()

plt.show()
```
